```python
#Adjusting the Dataframe for Stock Splits and Dividends if happend
import yfinance as yf
import pandas as pd

# Creating a Ticker object for Jio Platforms Limited (RELIANCE.NS)
ticker = yf.Ticker("RELIANCE.NS")

# Getting stock split and dividend actions
actions = ticker.actions

# Setting the date range for analysis (10 years ago to now)
start_date = "2013-11-01"
end_date = "2023-11-01"

# Creating a Ticker object for Jio Platforms Limited (RELIANCE.NS) with the specified date range
ticker = yf.Ticker("RELIANCE.NS")

# Getting historical stock prices within the specified date range
df = ticker.history(start=start_date, end=end_date)

# Initializing a new DataFrame df1 with the same data as df
df1 = df.copy()

# Initialize dictionaries to store the adjustment factors
split_factors = {}
dividend_factors = {}

# Iterating through actions to calculate adjustment factors
for action_date, action in actions.iterrows():
    action_date_str = action_date.strftime("%Y-%m-%d")

    # Handle stock splits
    if action["Stock Splits"] > 1:
        split_factors[action_date_str] = 1 / action["Stock Splits"]

    # Handle dividends
    if action["Dividends"] > 0:
        dividend_factors[action_date_str] = 1 - action["Dividends"]

# Adjusting the "Close" prices for stock splits and dividends
for date, row in df1.iterrows():
    date_str = date.strftime("%Y-%m-%d")

    if date_str in split_factors:
        df1.loc[date, "Close"] *= split_factors[date_str]

    if date_str in dividend_factors:
        df1.loc[date, "Close"] /= dividend_factors[date_str]

# Displaying the adjusted DataFrame df1
print(df1)
```
