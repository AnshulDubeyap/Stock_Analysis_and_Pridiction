```python
import pandas as pd
```

```python
df1 = pd.read_csv("jio_adjusted_stock_prices.csv")
```

```python
print(df1.head())
```

```python
import matplotlib.pyplot as plt

# Extracting the "Close" prices and "Date" column from your DataFrame
close_prices = df1['Close']
dates = df1['Date']

# Converting the "Date" column to a proper datetime format
dates = pd.to_datetime(dates)

# Creating a line chart to visualize the closing prices
plt.figure(figsize=(12, 6))
plt.plot(dates, close_prices, label='Closing Price', color='blue')
plt.title('Jio Platforms Limited (RELIANCE.NS) Closing Prices Over Time')
plt.xlabel('Date')
plt.ylabel('Price (INR)')
plt.legend()
plt.grid(True)

# Showing the plot
plt.show()
```

```python
#Insights from above plot
#1-Their is a overall growth in stock closing price
#2-the sudden dip and rebound in 2020-2021 can be due to various factors 
#2.1-Farmers' boycott. During the 20202021 Indian Farmers' Protest, farmers across multiple states in northern part of India boycotted and vandalized Jio's towers
#2.2-3dec2020 Reliance Jio Leads Series A Funding Of Mobile Gaming Startup Krikey. Mukesh Ambani's Reliance Jio made a foray into AI and gaming on Wednesday leading the Series A funding of mobile gaming company Krikey. The total funding raised by the San Francisco-based startup now stands at $22 Mn.
#2.3-Sep 2020  Partners with Cisco Systems for 5G deployment
#2.4-Sep 2020  Announces partnership with HFCL to deploy Fiber-optic communication to support the rollout of FTTx services
#3-their is a cyclic pattern in huge dip every year
```

```python
import matplotlib.pyplot as plt

# Extracting the "Volume" and "Date" columns from your DataFrame
volume = df1['Volume']
dates = df1['Date']

# Converting the "Date" column to a proper datetime format
dates = pd.to_datetime(dates)

# Creating a bar chart to visualize the trading volume
plt.figure(figsize=(12, 6))
plt.bar(dates, volume, color='green', label='Trading Volume')
plt.title('Jio Platforms Limited (RELIANCE.NS) Trading Volume Over Time')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.legend()
plt.grid(True)

# Showing the plot
plt.show()
```

```python
#Insights for above plot
#1-spike in 2017 is due to Reliance shares jump 10% after Jio announces tariff plans......On Tuesday, Reliance Jio announced that it will end free data service and start charging its customers from 1 April, though voice calls and roaming will remain free.
#2-spike in 2020-2021 
```

```python
#Moving Averages
# Calculating short-term and long-term moving averages
short_term_ma = close_prices.rolling(window=50).mean()
long_term_ma = close_prices.rolling(window=200).mean()

# Creating a plot to visualize the moving averages
plt.figure(figsize=(12, 6))
plt.plot(dates, close_prices, label='Closing Price', color='blue')
plt.plot(dates, short_term_ma, label='50-day Moving Average', color='orange')
plt.plot(dates, long_term_ma, label='200-day Moving Average', color='red')
plt.title('Moving Averages for Jio Platforms Limited (RELIANCE.NS)')
plt.xlabel('Date')
plt.ylabel('Price (INR)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Price Volatility
# Calculating Bollinger Bands
rolling_mean = close_prices.rolling(window=20).mean()
rolling_std = close_prices.rolling(window=20).std()
upper_band = rolling_mean + (rolling_std * 2)
lower_band = rolling_mean - (rolling_std * 2)

# Creating a plot to visualize Bollinger Bands
plt.figure(figsize=(12, 6))
plt.plot(dates, close_prices, label='Closing Price', color='blue')
plt.plot(dates, upper_band, label='Upper Bollinger Band', color='red')
plt.plot(dates, lower_band, label='Lower Bollinger Band', color='green')
plt.fill_between(dates, lower_band, upper_band, color='lightgray')
plt.title('Bollinger Bands for Jio Platforms Limited (RELIANCE.NS)')
plt.xlabel('Date')
plt.ylabel('Price (INR)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Price correlation
# Calculating the correlation matrix
# Convert the "Date" column to a datetime format
df1['Date'] = pd.to_datetime(df1['Date'])

# Selecting only numeric columns for correlation
import numpy as np
numeric_columns = df1.select_dtypes(include=[np.number])

# Calculating the correlation matrix
correlation_matrix = numeric_columns.corr()

# Creating a heatmap to visualize correlations
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()

```
